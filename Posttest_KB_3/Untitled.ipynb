## Import Pandas

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("vgsales.csv") 
df
```

# Perbandingan menggunakan seaborn
##### dalam grafik dibawah ini kita bisa mendapatkan kesimpulan bahwa action pada tahun 2008 merupakan jenis platform yang terbanyak


```python
dataplotting=df.sort_values("Name").head(30)
sns.set_style("ticks")
plt.figure(figsize = (8,9))
sns.countplot(data = dataplotting, x = "Genre", hue = "Year", palette = "rocket")
plt.title('Perbandingan Platfrom dengan Year')
```

## Korelasi antara 2 attribute yang bertipe numeric

```python
plt.figure(figsize=(10,5))

sns.heatmap(
    df.corr(),
    annot=True,
    cmap='flare'
) 
plt.title('korelasi antar attribut variabel ')

plt.show()
```

## Membuat Diagram Pie Pada Attribute Platform    
#### kita akan membuat bentuk data Platform ke dalam persenan,contohnya DS dengan jumlah 2163 mendapatkan hasil 13%

```python
data_pie = df["Platform"].value_counts()
data_pie.head()
```

```python
data_pie = df["Platform"].value_counts().rename_axis("Platform")
data_pie.head()
```

```python
data_pie = df["Platform"].value_counts().rename_axis("Platform").reset_index(name = "id_count")
```

```python


plt.pie(data_pie.id_count,
    labels = data_pie["id_count"], 
    startangle=90, 
    autopct='%.1f%%', 
    shadow=True
)
plt.title('diagram pie perbandingan Platform dalam sebuah Aplikasi')
plt.show()

```
