## membagi training set dan testing set dengan proporsi  70:30

## Data Cleaning

```python
from sklearn.impute import SimpleImputer
df.isna().sum()
```

## Mengubah nilai null yang punya tipe data desimal

```python
si = SimpleImputer(strategy = 'mean')
df["Global_Sales"] = si.fit_transform(df[["EU_Sales"]])
print("\nJumlah null setelah menggunakan SimpleImputer: ")
print(df.isna().sum())
```

## Mengubah nilai null yang punya tipe data kategorikal

```python
si = SimpleImputer(strategy = 'most_frequent')
df["Publisher"] = si.fit_transform(df[["Name"]])
print("\nJumlah null setelah menggunakan SimpleImputer: ")
print(df.isna().sum())
```

## Mengubah nilai null yang punya tipe data bilang bulat

```python
si = SimpleImputer(strategy = 'most_frequent')
df["Rank"] = si.fit_transform(df[["Rank"]])
print("\nJumlah null setelah menggunakan SimpleImputer: ")
print(df.isna().sum())
```
